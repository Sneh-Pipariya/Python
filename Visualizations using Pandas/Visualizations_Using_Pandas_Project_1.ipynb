```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading CSV file

```python
print(plt.style.available)
plt.style.use('ggplot')
```

```python
df = pd.read_csv("/Users/snehpipariya/Documents/Data Analyst Bootcamp by Alex Youtuber/Python/Visualizations using Pandas/Ice Cream Ratings.csv")
df = df.set_index("Date")
df
```

# LINE plot

```python
df.plot(kind="line",title="Ice Cream Ratings", xlabel="Daily Ratings", ylabel="Scores")
```

# BAR plot

```python
df.plot.barh(stacked=True)
```

# SCATTER plot

```python
df.plot.scatter(x="Flavor Rating", y="Overall Rating", s=200, c="Red")
```

# HISTOGRAM

```python
df.plot.hist(bins = 20)
```

# BOX plot

```python
df.boxplot()
```

# AREA plot

```python
df.plot.area(figsize = (10,5))
```

# PIE chart

```python
df.plot.pie(y='Flavor Rating', figsize = (10,9))
```
